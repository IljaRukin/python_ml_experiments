### Viola Jones Algorithm

with haarcascade_frontalface_default.xml from opencv (https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml)<br><br>

Example for Haar filters:

```python
%reset -f
import matplotlib.pyplot as plt
import cv2
import numpy as np

#load classifier and create detector
detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')

#image
#BGR_pic = cv2.imread('pictures/girl.png')
BGR_pic = cv2.imread('people.jpg')
pic = cv2.cvtColor(BGR_pic, cv2.COLOR_BGR2RGB)
bw_pic = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
```

```python
#face detection
faces = detector.detectMultiScale((bw_pic), scaleFactor=1.08, minNeighbors=12, minSize=(50,50))

#add boxes
frame=pic
for (x,y,w,h) in faces:
    pt1=(x,y)
    pt2=(x+w,y+h)
    frame = cv2.rectangle(frame, pt1, pt2, (0, 0, 255), 20)


#plot
fig = plt.figure(figsize=(10,8), dpi=200)

plt.subplot(1,1,1)
plt.imshow(pic)
plt.axis('off')
plt.show()
```
